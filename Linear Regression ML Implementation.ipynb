```python
## Independent and dependent features

X=dataset.iloc[:,:-1] # take all the columns and skip the "price"
y=dataset.iloc[:,-1] # take only "price" and skip all the columns
```

```python
X.head()
```

```python
y
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
## Every feature over here is calculated with respect to different different units , And to make 
## Gradient descent faster we have to standardize all these data points to the same scale/units 
## (because we have to reached the global minima for that necessory to make same scale,
## also we are using gradient descent inside the linear regression algorithm)
## And that should be based on something called the Standardize scaller.


## Standardize the dataset
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
X_train
```

## Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
## print the coefficient and the intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
## On which parameters the model has been trained
regression.get_params()
```

```python
## Prediction with test data
reg_pred=regression.predict(X_test)
```

```python
reg_pred
```

## Assumptions

```python
## plot a scatter plot for the prediction
plt.scatter(y_test,reg_pred) # "y_test" is truth value and "reg_pred" is predicted values
plt.show()  # when this ploting is basically linear so yes your model have actually perform well
```

```python
## Residuals means error between y_test and reg_pred
residuals=y_test-reg_pred
```

```python
residuals
```

```python
## Plot this residuals

sns.kdeplot(residuals)
plt.show()
```

```python
## Scatter plot with respect to prediction and residuals
## Uniform distribution --> gives the indication that your model is performing well
plt.scatter(reg_pred,residuals)
plt.show()
```

```python
## How my model is basically performing (performance metrics)

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_squared_error(y_test, reg_pred))
print(mean_absolute_error(y_test, reg_pred))
print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```

## R square and adjusted R square

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
```

### Adjusted R2 = 1-[(1-R2)*(n-1)(n-k-1)]

#### where: 
#### R2: The R2 of the model 
#### n: The number of the observations
#### k: The number of predicted variables

```python
# display adjusted R-squared
1 - (1-score)*(len(y_test) -1)/(len(y_test)-X_test.shape[1]-1)
```

## Pickling The Model file for Deployment

The pickle module implements binary protocols for serializing and de-serializing a Python object structure. Pickling is the process whereby a Python object hierarchy is converted into a byte stream, and unpickling is the inverse operation, whereby a byte stream (from a binary file or bytes-like object) is converted back into an object hierarchy. Pickling (and unpickling) is alternatively known as serialization, marshalling, 1 or flattening; however, to avoid confusion, the terms used here are pickling and unpickling.

Pickle in Python is primarily used in serializing and deserializing a Python object structure. In other words, it's the process of converting a Python object into a byte stream to store it in a file/database, maintain program state across sessions, or transport data over the network.

```python
import pickle
```

```python
pickle.dump(regression,open('regmodel.pkl','wb'))
```

```python
pickled_model=pickle.load(open('regmodel.pkl','rb'))
```

```python
## prediction 
pickled_model.predict(scaler.transform(boston.data.iloc[0].values.reshape(1, -1)))
```
